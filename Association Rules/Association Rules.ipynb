```python
import numpy as np
import pandas as pd
import pickle as pic
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import OnehotTransactions, TransactionEncoder
import plotly.graph_objects as go
import plotly as plt
import plotly.express as ex
```

```python
# Pre-Processing
# http://intelligentonlinetools.com/blog/2018/02/10/how-to-create-data-visualization-for-association-rules-in-data-mining/
data = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
  
# You don't need 'with' unless you're using the f object you created multiple times
# it just sets up the stream as an object within the scope of with
'''
with open() as f 

    for r in f.readlines():
        r is a line in the file # here the newline characters are already taken into account to separate the lines
        # in the first place
        my_list.append( r.split(' '))
        
    for r in f.read():
        r is all the lines # r includes the newline character to differentiates lines
''' 
    
data_file = open("T10I4D100K.dat", 'r')
data_T10 = data_file.read().replace(' ', '').replace('\n', '').split(',')
data_file.close()

data_file = open("mushroom.dat", 'r')
data_Mush = data_file.read().replace(' ', '').replace('\n', '').split(',')
data_file.close()

data_file = open("Teams.dat", 'r')
data_Team = data_file.read().replace(' ', '').replace('\n', '').split(',')
data_file.close()


for sup in range(1,4,1):
    oht = TransactionEncoder()
    df = pd.DataFrame( oht.fit(data_T10).transform(data_T10), columns = oht.columns_)
    frequent_itemsets = apriori(df, min_support = float(sup/100), use_colnames = True)
    print("T10I4D100K Dataset with Support: " + str(float(sup/100)))
    print(frequent_itemsets)
    
    rules_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
    rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
    print("T10I4D100K Dataset with Support: " + str(float(sup/100)) + " Association Rules with Confidence Metric")    
    print(rules_confidence)
    print("T10I4D100K Dataset with Support: " + str(float(sup/100)) + " Association Rules with Lift Metric")    
    print(rules_lift)

for sup in range(5,21,5):
    oht = TransactionEncoder()
    df = pd.DataFrame( oht.fit(data_Mush).transform(data_Mush), columns = oht.columns_)
    frequent_itemsets = apriori(df, min_support = float(sup/100), use_colnames = True)
    print("Mushroom Dataset with Support: " + str(float(sup/100)))
    print(frequent_itemsets)
    
    rules_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
    rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
    print("Mushroom Dataset with Support: " + str(float(sup/100)) + " Association Rules with Confidence Metric")    
    print(rules_confidence)
    print("Mushroom Dataset with Support: " + str(float(sup/100)) + " Association Rules with Lift Metric")    
    print(rules_lift)
    
for sup in range(5,101,5):    
    oht = TransactionEncoder()
    df = pd.DataFrame( oht.fit(data_Team).transform(data_Team), columns = oht.columns_)
    frequent_itemsets = apriori(df, min_support = float(sup/100), use_colnames = True)
    print("Team Dataset with Support: " + str(float(sup/100)))
    print(frequent_itemsets)
    
    rules_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
    rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
    print("Team Dataset with Support: " + str(float(sup/100)) + " Association Rules with Confidence Metric")    
    print(rules_confidence)
    print("Team Dataset with Support: " + str(float(sup/100)) + " Association Rules with Lift Metric")    
    print(rules_lift)
```
